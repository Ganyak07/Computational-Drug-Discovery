**Calculating Fingerprint Descriptors**

Fingerprint descriptors are representations of chemical compounds based on their structural features, often used in cheminformatics and computational drug discovery. These descriptors encode information about molecular topology, atom types, bond types, and other structural characteristics, allowing for similarity comparison and virtual screening of compounds.

There are various types of fingerprint descriptors, such as Morgan fingerprints, ECFP (Extended-Connectivity Fingerprints), and MACCS (Molecular ACCess System) keys. These descriptors are typically binary vectors, where each bit represents the presence or absence of a specific substructure or pattern within the molecule.

**Application in Drug Discovery:**
Fingerprint descriptors are widely used in virtual screening and similarity-based methods to identify potential drug candidates. By calculating fingerprint descriptors for compounds targeting Acetylcholinesterase (AChE), we can compare their structural similarities with known bioactive molecules and prioritize compounds with similar fingerprints for further experimental validation.

**Calculating PaDEL Descriptors**

PaDEL (Prediction of Activity and Descriptor Learning) descriptors are a comprehensive set of molecular descriptors computed using the PaDEL software. These descriptors capture various physicochemical, topological, and structural properties of chemical compounds, providing valuable insights into their molecular characteristics.

PaDEL descriptors include a wide range of features, such as constitutional descriptors (e.g., molecular weight, number of atoms), topological descriptors (e.g., connectivity indices, Wiener index), and geometric descriptors (e.g., molecular surface area, volume). Additionally, PaDEL calculates substructure-based fingerprints and pharmacophore features, enhancing the richness of descriptor information.

**Application in Drug Discovery:**
PaDEL descriptors offer a holistic view of chemical compounds, facilitating the exploration of structure-activity relationships (SAR) and the prediction of biological activities. By calculating PaDEL descriptors for compounds targeting AChE, we can uncover key molecular features associated with their bioactivity and gain insights into their potential mechanisms of action. These descriptors serve as valuable inputs for machine learning models and quantitative structure-activity relationship (QSAR) analyses, guiding the design and optimization of novel drug candidates for Alzheimer's disease treatment.

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('acetylcholinesterase_06_bioactivity_data_3class_pIC50_pubchem_fp.csv')
```

```python
# Input features
X = df.drop('pIC50', axis=1)
X
```

```python
#output features
Y = df.pIC50
Y
```

```python
# X dimension
X.shape
```

```python
# y dimension
Y.shape
```

```python
# Removal of low varience
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
X = selection.fit_transform(X)
```

```python
X.shape
```

```python
# Spliting the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
X_train.shape, Y_train.shape
```

```python
#build the regression model using RandomForest
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
r2
```

```python
Y_pred = model.predict(X_test)
```

# **Scatter Plot of Experimental vs Predicted pIC50 Values**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(color_codes=True)
sns.set_style("white")

# Make sure Y_test and Y_pred are defined correctly before using them here
ax = sns.regplot(x=Y_pred, y=Y_test, scatter_kws={'alpha':0.4})
ax.set_xlabel('Predicted pIC50', fontsize='large', fontweight='bold')
ax.set_ylabel('Experimental pIC50', fontsize='large', fontweight='bold')
ax.set_xlim(0, 12)
ax.set_ylim(0, 12)
ax.figure.set_size_inches(5, 5)
plt.show()
```
